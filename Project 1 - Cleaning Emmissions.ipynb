```python
# Cleaning Emmissions CSV File

# Reading Emissions Data CSV & Preview
import pandas as pd
Emissions_dataframe = pd.read_csv("Europe Emissions by Country_2023.csv")
Emissions_dataframe.tail()
```

```python
emissions_summary = Emissions_dataframe[['Country name','Gas/Scope','Emissions','Description test']]
emissions_summary = emissions_summary.rename(columns={"Country name": "Country Name", 
                                    "Gas/Scope": "Gas Type","Description test":"Emissions Activity Description"})
emissions_summary = emissions_summary.dropna(how='any')
emissions_summary.tail()
```
